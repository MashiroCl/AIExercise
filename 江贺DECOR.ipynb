```python
filename=r'E:\江贺项目\1_27\Hsqldb.txt'
```

```python
with open(filename) as f:
    text_lines=f.readlines()

text=[]
for i in range(len(text_lines)):
    if(i+6<len(text_lines) and text_lines[i].find("------>")!=-1):
        text.append(text_lines[i+6].split("=")[1].split("\n")[0].split()[0]) 
text
```

```python
import csv
csv_Validation=r"E:\江贺项目\12_19\validation_exp\Hsqldb.csv"
data=csv.reader(open(csv_Validation))
new_csvV=[]
temp_num=0
for row in data:
    temp=[]
    temp.append(row[1])
    temp.append(row[34])
    new_csvV.append(temp)
new_csvV
```

```python
y_DECOR=[]
y_Validation=[]
for eachV in new_csvV[1:]: #去掉表格第一行的['Address', 'LABEL']
    flag=0
    if(eachV[1]=='0'):
        y_Validation.append(0)
    elif(eachV[1]=='1'):
        y_Validation.append(1)
    for eachD in text[1:]: 
        if(eachD==eachV[0] and flag==0):
            y_DECOR.append(1)
            flag=1
    if(flag==0):
        y_DECOR.append(0)
# y_Validation[0],y_DECOR[0]
len(y_Validation),len(y_DECOR)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_Validation,y_DECOR))
```
